```python
def VarStd(dataset,quan):
    Var_Std=pd.DataFrame(index=['Variance','Std_Deviation'],columns=quan)
    for columnName in quan:
        Var_Std[columnName]['Variance']=dataset[columnName].var()
        Var_Std[columnName]['Std_Deviation']=dataset[columnName].std()
    return Var_Std
```

```python
table=VarStd(dataset,quan)
table
```
