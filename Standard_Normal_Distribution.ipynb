```python
import pandas as pd
```

```python
dataset=pd.read_csv("Preprocessed_Placement.csv")
```

```python

def SNDgraph(dataset):
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    value=[i for i in dataset]
    z_score=[((x-mean)/std) for x in value]
    sns.distplot(z_score,kde=True,kde_kws={'color':'red'},color='blue')
    s=sum(z_score)/len(z_score)
```

```python
SNDgraph(dataset['ssc_p'])
```
